# Projectile Motion
<center><img src="Projectile Motion.png"></center>

### In Horizontal Direction:

> $Displacement = \Delta x = (v_o cos \theta) t$ 

> $velocity = v_x = v_o cos \theta$

### In Vertical Direction:

> $Displacement = \Delta y = \frac{1}{2} (-g)t^2 + (v_o sin \theta) t$

> $velocity = v_y = v_o sin \theta$

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi,sin,cos

u0 = 40    #projectile velocity (m/s)
g = 9.81   #acceleration due to gravity (m/s^2)

theta = 45 #projectile angle

# deg to rad conversion: deg * pi/180
ux = u0*cos(theta * (pi/180))
uy = u0*sin(theta * (pi/180))

# projectile total time:
t_total = 2*uy/g

t = np.linspace(0,t_total,100)

# projectile displacements with respect to time:
sx = ux*t
sy = (uy*t) - (0.5*g*t**2)

plt.figure(figsize=(10,5))
plt.plot(sx,sy,label="theta = 45")
plt.ylim(0,50)
plt.show()
```
