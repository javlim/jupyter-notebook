```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi,sin,cos

Syt = 200  # Yield Strength
s1 = 150   # Stress in direction 1
s2 = 80    # Stress in direction 2

Sv = np.sqrt(s1**2 - (s1*s2) + s2**2)

a = 2**0.5*Syt # semi major axis of ellipse
b = (2/3)**0.5*Syt #semi minor axis of ellipse

theta = np.linspace(0,360*pi/180,720)

# orignal points (x,y)
x = a*cos(theta)
y = b*sin(theta)

# rotation angle
rot_ang = pi/4

xd = x*cos(rot_ang) - y*sin(rot_ang)
yd = x*sin(rot_ang) + y*cos(rot_ang)

# stress points
xp = [Syt,Syt,0,-Syt,-Syt,0,Syt]
yp = [0,Syt,Syt,0,-Syt,-Syt,0]

plt.figure(figsize=(7.5,5))
plt.plot(xd,yd,label="Von Mises Region")
plt.plot(xp,yp,linestyle="--",label="Max Shear Region")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi,sin,cos

Syt = 200  # Yield Strength
s1 = 150   # Stress in direction 1
s2 = 80    # Stress in direction 2

Sv = np.sqrt(s1**2 - (s1*s2) + s2**2)

a = 2**0.5*Syt # semi major axis of ellipse
b = (2/3)**0.5*Syt #semi minor axis of ellipse

theta = np.linspace(0,360*pi/180,720)

# orignal points (x,y)
x = a*cos(theta)
y = b*sin(theta)

# rotation angle
rot_ang = pi/4

xd = x*cos(rot_ang) - y*sin(rot_ang)
yd = x*sin(rot_ang) + y*cos(rot_ang)

# stress points
xp = [Syt,Syt,0,-Syt,-Syt,0,Syt]
yp = [0,Syt,Syt,0,-Syt,-Syt,0]

# Sample Stress
Sp1 = [200,160,80,60]
Sp2 = [-80,210,50,40]

plt.figure(figsize=(7.5,5))
plt.plot(xd,yd,label="Von Mises Region")
plt.plot(xp,yp,linestyle="--",label="Max Shear Region")
plt.scatter(Sp1,Sp2)
plt.legend()
plt.show()
```
